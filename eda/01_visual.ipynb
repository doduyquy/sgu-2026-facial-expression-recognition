```python
import random
import os
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# 1. Cấu hình đường dẫn (Thay đổi cho đúng máy của bạn)
data_dir = r'D:\HocTap\Phân tích  và xử lý ảnh\sgu-2026-facial-expression-recognition\data\fer-2013_image\train'

# 2. Lấy danh sách tên thư mục cảm xúc (angry, disgust, fear, v.v.)
emotions = sorted([d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))])
```

```python
plt.figure(figsize=(12, 18))
plot_index = 1

for mood in emotions:
    label_path = os.path.join(data_dir, mood)
    
    # Lấy danh sách file ảnh trong thư mục cảm xúc đó
    images = [f for f in os.listdir(label_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    if len(images) > 0:
        # Lấy ngẫu nhiên 3 ảnh để hiển thị mỗi hàng
        sample_images = random.sample(images, min(len(images), 3))
        
        for img_name in sample_images:
            img_path = os.path.join(label_path, img_name)
            img = Image.open(img_path).convert("RGB")
            
            # Tạo subplot (7 hàng tương ứng 7 cảm xúc, 3 cột tương ứng 3 ảnh mẫu)
            plt.subplot(7, 3, plot_index)
            plt.imshow(img)
            plt.title(mood)
            plt.axis("off")
            
            # QUAN TRỌNG: Phải dùng += để tăng vị trí vẽ
            plot_index += 1 

plt.tight_layout()
plt.show()
```
