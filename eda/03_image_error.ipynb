```python
import os
import numpy as np
from PIL import Image, ImageStat
from collections import Counter

# Đường dẫn đến thư mục 'train' hoặc 'test' của bạn
image_dir = r"D:\HocTap\Phan tich va xu ly anh\sgu-2026-facial-expression-recognition\data\fer-2013_image\train"
```

```python
shapes = []
brightness_list = []
low_brightness_files = [] # Lưu các ảnh quá tối (< 20)
high_brightness_files = [] # Lưu các ảnh quá sáng (> 230)
```

```python
for root, dirs, files in os.walk(image_dir):
    for img_name in files:
        if img_name.lower().endswith(('.jpg', '.png', '.jpeg')):
            img_path = os.path.join(root, img_name)
            
            try:
                img = Image.open(img_path).convert('L') # Convert sang ảnh xám để tính độ sáng
                
                # 1. Thống kê kích thước
                shapes.append(img.size)
                
                # 2. Tính độ sáng trung bình (0 là đen, 255 là trắng)
                stat = ImageStat.Stat(img)
                avg_brightness = stat.mean[0]
                brightness_list.append(avg_brightness)
                
                # Kiểm tra ảnh lỗi độ sáng
                if avg_brightness < 20:
                    low_brightness_files.append(img_path)
                elif avg_brightness > 230:
                    high_brightness_files.append(img_path)
                    
            except Exception as e:
                print(f"Lỗi đọc ảnh {img_name}: {e}")
```

```python
# --- Kết quả thống kê ---
print("-" * 30)
print("1. THỐNG KÊ KÍCH THƯỚC (SHAPE):")
shape_count = Counter(shapes)
for size, count in shape_count.items():
    print(f" - Kích thước {size}: {count} ảnh")

print("\n2. THỐNG KÊ ĐỘ SÁNG (BRIGHTNESS):")
print(f" - Độ sáng TB toàn bộ: {np.mean(brightness_list):.2f}")
print(f" - Số ảnh quá tối (cần check): {len(low_brightness_files)}")
print(f" - Số ảnh quá sáng (cần check): {len(high_brightness_files)}")
print("-" * 30)
```

```python
import matplotlib.pyplot as plt

if low_brightness_files:
    plt.figure(figsize=(10, 3))
    print("Ví dụ các ảnh quá tối:")
    for i, path in enumerate(low_brightness_files[:5]): # Xem thử 5 tấm đầu
        plt.subplot(1, 5, i+1)
        img = Image.open(path)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
    plt.show()
```
