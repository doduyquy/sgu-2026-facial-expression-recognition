```python
import pandas as pd

# =========================
# CONFIG
# =========================
CSV_PATH = "../../data/cleaned_data/manual_remove_candidates.csv"   # file 3 cột của bạn
OUTPUT_PATH = "../../data/cleaned_data/manual_remove_candidates_cleaned.csv"

# Danh sách row_id muốn xóa khỏi file này
# row_ids_to_delete = [20417 , 4102, 5023, 5534, 6716, 7497, 11800, 12775, 13923, 14476, 15655, 17853, 19176, 26624]
row_ids_to_delete = [837, 1152]

# =========================
# LOAD
# =========================
df = pd.read_csv(CSV_PATH)

required_cols = {"row_id", "emotion", "Usage"}
missing = required_cols - set(df.columns)
if missing:
    raise ValueError(f"Thiếu cột: {missing}")

print("Số dòng trước khi xóa:", len(df))

# =========================
# CLEAN
# =========================
row_ids_to_delete = set(row_ids_to_delete)

df_cleaned = df[~df["row_id"].isin(row_ids_to_delete)].copy()

print("Số dòng cần xóa:", len(df) - len(df_cleaned))
print("Số dòng sau khi xóa:", len(df_cleaned))

# =========================
# SAVE
# =========================
df_cleaned.to_csv(OUTPUT_PATH, index=False)
print("Đã lưu file:", OUTPUT_PATH)

display(df_cleaned.head())
```
